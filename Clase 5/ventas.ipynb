Un script que pregunta un intervalo de años, y luego pregunta el nivel de ventas de cada uno de esos años, y luego hace un gráfico x-y de la evolución de las ventas

```python
import matplotlib.pyplot as plt
```

```python
def obtener_año(mensaje = "Ingrese el año: "):
    while True:
        try:
            año = int(input(mensaje))
            if año > 0:
                return año
            else:
                print("El año debe ser número positivo.")
        except ValueError:
            print("Entrada no válida. Por favor, ingrese un número entero positivo.")

inicio = obtener_año("Ingrese el año de inicio: ")
fin = obtener_año("Ingrese el año de fin: ")
ventas = {}
for i in range(inicio, fin + 1):
    while True:
        try:
            venta = float(input(f"Ingrese las ventas del año {i}: "))
            if venta >= 0:
                ventas[i] = venta
                break
            else:
                print("Las ventas deben ser un número positivo.")
        except ValueError:
            print("Entrada no válida. Por favor, ingrese un número decimal positivo.")
fig, ax = plt.subplots()
ax.plot(ventas.keys(), ventas.values() , marker='o')
ax.set_title('Evolución de ventas por año')	
```
