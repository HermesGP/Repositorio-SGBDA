Importación de librerías

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Creación del dataframe de películas

```python
df = pd.read_excel("Netflix-Movies-Sample-Data.xlsx", skiprows=4, header=1) #Omitiendo las primeras 4 filas y la primera fila sera el encabezado
df = df.drop(columns=["Unnamed: 0"]) #Eliminando la columna sin nombre
df.head(10)
```

```python
df_movie_durations = df["Duration"].str.extract(r"(\d+)").astype(int) # Extrayendo la duración de las películas y convirtiéndola a entero
# Graficando un histograma de las duraciones de las películas
df_movie_durations.plot(kind="hist", bins=20, color="red", edgecolor="black")
plt.title("Distribución de duraciones de películas")
plt.xlabel("Duración (minutos)")
```

```python
df_imdb_ratings = df["IMDb Rating"].astype(float)
print(df_imdb_ratings.describe())
# Graficando un histograma de las calificaciones IMDB
df_imdb_ratings.plot(kind="hist", bins=10, color="lightblue", edgecolor="black")
```

```python
df_age_ratings = df["Age Rating"].value_counts()
df_age_ratings.plot(kind="pie")
plt.title("Distribución de calificaciones por edad")
```

```python
df_decades = pd.cut(df["Year"], bins = range(1920, 2030, 10), right=False)
df_decades = df_decades.value_counts().sort_index()
df_decades.plot(kind="bar", color="green", edgecolor="black")
plt.title("Distribución de películas por década")
```
