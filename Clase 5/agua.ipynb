```python
# importando las librerias necesarias
import pandas as pd
import matplotlib.pyplot as plt
```

Creación del dataframe

```python
# Cargamos el archivo CSV
# con el separador ';' y lo guardamos en un dataframe
df_agua = pd.read_csv('poblacionconagua.csv', sep=';')
df_agua
```

Preparación de los datos para los gráficos

```python
df_agua = df_agua.drop(columns=["Ciudad"]) # Eliminamos la columna Ciudad
df_agua_metro = df_agua.iloc[1].astype('float') # Creamos un dataframe de la fila 1 (con los valores para Gran Asunción) con el tipo float para poder graficar
print(df_agua_metro.head())
df_agua_interior = df_agua.iloc[10].astype('float')# Lo mismo para la fila 10 del dataframe (con los valores para el interior del país)
print(df_agua_interior.head())
```

Graficamos

```python
df_agua_metro.plot(title='Evolución de la población con agua en la zona metropolitana')
plt.ylabel('Población con agua en miles de habitantes')
plt.xlabel('Año')
```

```python
df_agua_interior.plot(title='Evolución de la población con agua en el interior')
plt.ylabel('Población con agua en miles de habitantes')
plt.xlabel('Año')
```

En un solo gráfico

```python
y1 = df_agua_metro
y2 = df_agua_interior
plt.plot(y1, label='Zona Metropolitana')
plt.plot(y2, label='Interior')
```
